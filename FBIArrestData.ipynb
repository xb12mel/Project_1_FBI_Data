```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
from pprint import pprint
from api_keys import api_key
```

```python
base_url = "https://api.usa.gov/crime/fbi/sapi/api/"
```

```python
# NATIONAL LEVEL arrest-data-controller (National level Arrest Demographic Count Endpoint)

# Setting variable for offense so that you can easily have the URL filter by specific offense type
nat_since = "2000"
nat_until = "2019"

# This field can be all or monthly
nat_variable = "all"

# Building query URL
nat_query_url = f"{base_url}data/arrest/national/offense/{nat_variable}/{nat_since}/{nat_until}?API_KEY={api_key}"
nat_query_url
```

```python
# Getting data
nat_fbi_response = requests.get(nat_query_url)
nat_fbi_json = nat_fbi_response.json()
pprint(nat_fbi_json)
```

```python
# Testing if keys are correct to grab the specific data
print(nat_fbi_json["results"][0]["data_year"])
```

```python
# Creating lists to then later append with data from JSON data
nat_aggravated_assault = []
nat_year = []
nat_num = 0
for x in nat_fbi_json["results"]:
    nat_year.append(nat_fbi_json["results"][nat_num]["data_year"])
    nat_aggravated_assault.append(nat_fbi_json["results"][nat_num]["aggravated_assault"])
    nat_num +=1
```

```python
# Create a dictionary to hold the keys(titles) and lists
nat_fbi_dict = {
    "Data Year": nat_year,
    "Aggravated Assualt": nat_aggravated_assault 
}

nat_fbi_df = pd.DataFrame(nat_fbi_dict)
nat_fbi_df
```

```python
# Sorting new data frame by year
nat_sorted_df = nat_fbi_df.sort_values(by=["Data Year"])
nat_sorted_df
```
